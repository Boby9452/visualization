```python
import matplotlib.pyplot as plt
import seaborn as snr
import pandas as pd
```

```python
my_data=pd.read_csv('/content/FitBit data.csv')
```

```python
my_data.tail()
```

# SCATTER PLOT

```python
my_data.plot.scatter(x='TotalSteps',y='TotalDistance',c='red')
```

This graph shows the data of watch that how much distance and steps are covered by peoples.For a man covering distance more than 25km will travel steps more than 25000.

# RELATIVE PLOT

```python
snr.relplot(x='TrackerDistance',y='TotalSteps',data=my_data,size=10)
```

This graph helps to find the value of total steps covered corresponding to tracker distance.For ex :- a man has travel 10000 steps has tracker distance of approx 1.

# CATPLOT

```python
snr.catplot(x="LightActiveDistance",y='VeryActiveDistance',data=my_data)
```

```python
snr.catplot(x='ModeratelyActiveDistance',y='SedentaryActiveDistance',data=my_data)
```

# BOX PLOT

```python
snr.catplot(x='SedentaryActiveDistance',y='ModeratelyActiveDistance',data=my_data,kind='box')
```

# PAIRPLOT

```python
snr.pairplot(x_vars=['ModeratelyActiveDistance','LightActiveDistance'],y_vars='SedentaryActiveDistance',data=my_data)
```

It shows sedentaryactivedistance corresonding to moderatelyactivedistance and lightactivedistance.

# VIOLIN

```python
snr.catplot(x='ModeratelyActiveDistance',y='LightActiveDistance',data=my_data,kind='violin')
```

# STRIP

```python
snr.catplot(x='ModeratelyActiveDistance',y='LightActiveDistance',data=my_data,kind='strip')
```
