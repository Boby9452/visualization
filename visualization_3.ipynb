```python
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
titanic=snr.load_dataset('titanic')
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.plot.scatter(x='age',y='sex')
```

```python
snr.catplot(x='age',y='pclass',hue='sex',col='survived',data=titanic)
```

By observing this graph we can say that graph having survived=0 has more numbers of male peoples while survived=1 has more numbers of females. 

```python
snr.catplot(x='age',y='pclass',kind='box',data=titanic)
```

```python
snr.catplot(x='sex',y='age',kind='box',data=titanic)
```

Highest age of male is approximately 65 and highest age of female is approximately 62 and both box plot is in balance condition.

```python
snr.pairplot(data=titanic,x_vars=['pclass','fare'],y_vars='age')
```

By observing this graph we conclude that only one person has fare more than 400 whose age may be around 35.
Mostly the fare for all peoples is till 250.
pclass=1.0 has highest age people of 80.
